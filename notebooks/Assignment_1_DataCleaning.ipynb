```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Library successfully imported")
```

```python
file_path='/content/EmergencyTriageDataset_Reduced_Dirty.csv'
df_raw=pd.read_csv(file_path)
print(f"Data set loaded:{df_raw.shape[0]} rows x {df_raw.shape[1]} columns")
print(f"Columns:{list(df_raw.columns)}")
df_raw.head(15)
```

```python
print(df_raw.dtypes)
print("Unique gender values")
print(df_raw['Gender'].unique())
print(f"\n Total unique values:{df_raw['Gender'].nunique()}")
```

```python
print("Gender value counts:")
print(df_raw['Gender'].value_counts())

# Normalize dats (convert text to lowercase and remove extra spaces for consistency)

df_raw['Gender'] = df_raw['Gender'].astype(str).str.strip().str.lower()
gender_map = {
    'male':   1,
    '1':      1,
    'm':      1,
    'female': 0,
    '0':      0,
    'f':      0,
}

df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map)
print("After cleaning:")
print(df_raw['Gender_Clean'].value_counts())
print(f"\nAny NaN values? {df_raw['Gender_Clean'].isnull().sum()}")
df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_Clean': 'Gender'})

print("Column 'Gender' after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtype}")
df_raw.head()
```
